<h3>1)Match Results<h3>

```python
#Importing the match result file 
with open('C:/Users/dell/Downloads/t20_json_files/t20_wc_match_results.json') as a:
    data1=json.load(a)
df_match_results=pd.DataFrame(data1[0]['matchSummary'])
df_match_results.head()
```

```python
#Renaming Scorecard as MatchId to derive realtion in Power Bi
df_match_results.rename({'scorecard':'MatchID'},axis=1,inplace=True)
df_match_results.head()
```

```python
#Create a match ids dictionary that maps team names to a unique match id
match_ids_dict = {}
for index, row in df_match_results.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    match_ids_dict[key1] = row['MatchID']
    match_ids_dict[key2] = row['MatchID']
```

```python
df_match_results.to_csv('C:/Users/dell/Downloads/t20_csv_files/Match_Summary.csv', index = False)
```

<h3>2)Batting Sumarry<h3>

```python
#Importing the Player Batting Summary file 
with open('C:/Users/dell/Downloads/t20_json_files/t20_wc_batting_summary.json') as b:
    data2=json.load(b)
all_batting_records=[]
for rec in data2:
    all_batting_records.extend(rec['battingSummary'])
df_batting=pd.DataFrame(all_batting_records)
df_batting.tail()
```

```python
#Creating a new column for out and not-out
df_batting['Out/Not-Out']=df_batting.dismissal.apply(lambda x:'Out' if len(x)>0 else "Not Out")
df_batting.head(11)
```

```python
#Removing The dismissal column
df_batting.drop(columns=["dismissal"],inplace= True)
df_batting.head(11)
```

```python
#Replacing Special Charecters
df_batting['batsmanName'] = df_batting['batsmanName'].str.replace('â€', '', regex=False)
df_batting
```

```python
#Creating MatchID column and maping it
df_batting['MatchID']= df_batting['match'].map(match_ids_dict)
df_batting
```

```python
df_batting.to_csv('C:/Users/dell/Downloads/t20_csv_files/Batting_Summary.csv', index = False)
```
